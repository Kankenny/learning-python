# Chaotic Lorenz Model 


<font color='gray' size="4">Instructions</font> 

1) Create a definition of this function:
dx/dt = s*(y - x)
dy/dt = r*x - y - x*z
dz/dt = x*y - b*z
dt/dt = 1.  (as with the example we went over today).
where s=10., r=28., and b=2.667.

Start with the initialConditions array for [x, y, z, t] = [0.0, 0.1, 1.05, 0.0].  (This is the set of values at the start -- the beginning of integration.)

Use a step size of 0.01, and integrate over 50 steps.

So, this is aimed to get you used to giving odeint all the information it needs to integrate these equations: initial conditions, values of parameters, step size to use, and how many steps to take.  It's a lot.  Take your time.  Don't hurry.

2) Then plot: (x,t), (y,t), and (z,t).

3) Finally, try a 3-dimensional plot of (x,y,z.)



<font color='gray' size="4">Libraries and Packages</font> 

```python
import scipy as sp
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D as ax
```

<font color='gray' size="4">Array Definitions</font> 
- start value: 0
- end value: 50
- step value: 0.01

```python
start = -50.
end = 50.
step = .01

times = np.arange(start, end, step)
initialState = [0.0, 0.1, 1.05, 0.0]
```

<font color='gray' size="4">Equation Definitions</font> 

```python
# Sinai, Ruelle and Bowen (Chaotic Attractors)

def chaoticLorenzModelEquations(initialState, times):
    s, r, b = 10., 28., 2.667
    
    [x, y, z, t] = initialState
    
    dxdt = s*(y - x)
    dydt = r*x - y - x*z
    dzdt = x*y - b*z
    dtdt = 1.
     
    return dxdt, dydt, dzdt, dtdt


integratedStates = odeint(chaoticLorenzModelEquations, initialState, times)
```

<font color='gray' size="4">Generate Different Plots</font> 
##### (x, t)
##### (y, t)
##### (z, t)
##### (x, y, z)

```python
# Plot: (x, t)

fig = plt.figure(1)
plt.subplot(1, 1, 1) # ~ plt.subplot(111) where param 1, 2, 3 = nRow, nCol, idx
plt.xlabel('T')
plt.title('X as a function of T', fontweight="bold")
plt.ylabel('X')
plt.plot(times, integratedStates[:, 0], 'r')
plt.show()
```

```python
# Plot: (y, t)

fig = plt.figure(1)
plt.subplot(1, 1, 1) # ~ plt.subplot(111) where param 1, 2, 3 = nRow, nCol, idx
plt.xlabel('T')
plt.title('Y as a function of T', fontweight="bold")
plt.ylabel('Y')
plt.plot(times, integratedStates[:, 1], 'b')
plt.show()
```

```python
# Plot: (z, t)

fig = plt.figure(1)
plt.subplot(1, 1, 1) # ~ plt.subplot(111) where param 1, 2, 3 = nRow, nCol, idx
plt.xlabel('T')
plt.title('Z as a function of T', fontweight="bold")
plt.ylabel('Y')
plt.plot(times, integratedStates[:, 2], 'y')
plt.show()
```

```python
# Plot: (x, y, z) 

fig = plt.figure(1)
ax = fig.gca(projection="3d", label="(X, Y, Z)")
ax.plot(integratedStates[:, 0], integratedStates[:, 1], integratedStates[:, 2], 'g')
ax.set_xlabel('X', fontsize=14, fontweight="bold")
ax.set_ylabel('Y', fontsize=14, fontweight="bold")
ax.set_zlabel('Z', fontsize=14, fontweight="bold")
plt.show()
```
